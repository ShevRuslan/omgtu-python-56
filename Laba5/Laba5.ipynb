## 2.9. создание нового поля вычисленного на основе значений других полей:

### 2.9.1 через выражение на базе имеющихся колонок,

```python
df['age'] = 2023 - df['year']
df
```

### 2.9.2 через DataFrame.apply

```python
df['age'] = (2023 - df['year']).apply(lambda x: max(x, 0)) #Функция lambda x: max(x, 0) возвращает значение x, если x
#больше или равно 0, и 0 в противном случае. Это нужно для того, чтобы избежать отрицательных значений возраста.
df
```

### 2.9.3 через Series.apply

```python
def calculate_age(year):
    return 2023 - year

df['age'] = df['year'].apply(calculate_age)
df
```
